```python
import requests
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
print('--'*15,end ='');print('preparing dataset from api',end ='');print('--'*15)
datau = requests.get("https://api.covid19india.org/data.json")
datau = datau.json()

dailyconfirmed = []
dailydeceased = []
dailyrecovered = []
date = []
totalconfirmed = []
totaldeceased = []
totalrecovered = []

df = pd.DataFrame(columns = ["date", "dailyconfirmed",'dailydeceased','dailyrecovered','totalconfirmed', 'totaldeceased', 'totalrecovered'])
for i in range(len(datau['cases_time_series'])):
    dailyconfirmed.append(datau['cases_time_series'][i]['dailyconfirmed'])
    dailydeceased.append(datau['cases_time_series'][i]['dailydeceased'])
    dailyrecovered.append(datau['cases_time_series'][i]['dailyrecovered'])
    date.append(datau['cases_time_series'][i]['date'])
    totalconfirmed.append(datau['cases_time_series'][i]['totalconfirmed'])
    totaldeceased.append(datau['cases_time_series'][i]['totaldeceased'])
    totalrecovered.append(datau['cases_time_series'][i]['totalrecovered'])

df['dailyconfirmed'] = dailyconfirmed
df['dailydeceased'] = dailydeceased
df['dailyrecovered'] = dailyrecovered
df['date'] = date
df['totalconfirmed'] =totalconfirmed
df['totalrecovered'] = totalrecovered
df['totaldeceased'] = totaldeceased
df['sl'] = df.index

df[["dailyconfirmed",'dailydeceased','dailyrecovered','totalconfirmed', 'totaldeceased', 'totalrecovered']] = df[["dailyconfirmed",'dailydeceased','dailyrecovered','totalconfirmed', 'totaldeceased', 'totalrecovered']].apply(pd.to_numeric)


df.to_csv("corona.csv")
print('--'*15,end ='');print("data stored in 'corona.csv'",end ='');print('--'*15)

# idx = pd.date_range('2020-01-30', periods= len(datau['cases_time_series']) , freq='D')
# # ts = pd.Series(range(len(idx)), index=idx)
# df['idx'] = idx

# df.set_index('idx',inplace = True)
```

```python
# prepare data
x1 = np.array(df["sl"]).reshape(-1,1)
y = np.array(df['totalconfirmed']).reshape(-1,1)
```

```python
print('--'*15,end ='');print('polynomial model training',end ='');print('--'*15)

for i in range(1,6):
    polyfet = PolynomialFeatures(degree=i)
    xa = polyfet.fit_transform(x1)
    model = linear_model.LinearRegression()
    model.fit(xa,y)
    accuracy = model.score(xa,y)
    print('accuracy(R2) with degree_{} is -->  {}%'.format(i , round(accuracy*100,3)))
print('--'*45)
```

```python
polyfet = PolynomialFeatures(degree=3)
# x1[:-1]
# y[:-1]
xa = polyfet.fit_transform(x1)
model = linear_model.LinearRegression()
model.fit(xa,y)
yp = model.predict(xa)
yact = np.array(df['totalconfirmed'])#.reshape(-1,1)
```

```python
plt.figure(figsize=(8, 6)) 
plt.plot(yp,"--b")
plt.plot(yact,"-g")
plt.legend(['pred', 'actual'])
plt.xticks()
# plt.yticks([])
plt.title("comparing actual and pred", fontdict=None, loc='center')
plt.show()
```

```python
#prediction
days = 30
print("Cases after {} day - ".format(days), end='')
print(round(int(model.predict(polyfet.transform(np.array(x1[-1:]+days).reshape(-1,1)))),2))
```
